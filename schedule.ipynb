```python
import pandas as pd
import numpy as np
import math
```

```python
schedule = pd.read_csv("sch.csv")
```

```python
teamData = pd.read_csv("rankings.csv")
teamData = teamData.drop(columns=["Weighted Score", "Avg Points", "Max Points", "Scout Flagged Matches","# Chargers","Drive Train","Frame Dimensions","Robustness Rating (0-5)", "Endgame Park %"])
teamData["Climb and Trap%"] = teamData["Climb and Trap%"] != "0.0%"
teamData["Climb OnStage %"] = (teamData["Climb OnStage %"] != "0.0%") | (teamData["Climb Spotlight %"] != "0.0%") | (teamData["Climb Harmony %"] != "0.0%") | (teamData["Climb and Trap%"] != False)
teamData = teamData.drop(columns=["Climb Spotlight %", "Climb Harmony %"])
```

```python
series = pd.Series([0] * teamData.shape[0], index=teamData["Team"])
rp = series.to_dict()
rp
```

```python
teamPos = pd.read_csv("teams.csv")
np.random.shuffle(teamPos["Pos"])
teamPos = teamPos.sort_values("Pos")
teamPos = teamPos.merge(teamData, how='inner', on="Team").set_index("Team")
teamPos
```

```python
for j in range(0,1): #Number of Simulations


    team1_arr = [""] * schedule.shape[0]
    team2_arr = [""] * schedule.shape[0]
    team3_arr = [""] * schedule.shape[0]
    team4_arr = [""] * schedule.shape[0]
    team5_arr = [""] * schedule.shape[0]
    team6_arr = [""] * schedule.shape[0]
    blueScoreArr = [0] * schedule.shape[0]
    redScoreArr = [0] * schedule.shape[0]
    blueRPArr = [0] * schedule.shape[0]
    redRPArr = [0] * schedule.shape[0]

    for i in range(0, schedule.shape[0]):
        team1_auto = teamPos.loc[schedule["team1"].iloc[i]]["Avg Auto Pieces"]
        team1_tele_speaker = teamPos.loc[schedule["team1"].iloc[i]]["Avg Teleop Speaker"]
        team1_tele_amp = teamPos.loc[schedule["team1"].iloc[i]]["Avg Teleop Amp"]
        team1_shuttle = teamPos.loc[schedule["team1"].iloc[i]]["Avg Shuttle"]
        team1_climb = teamPos.loc[schedule["team1"].iloc[i]]["Climb OnStage %"]
        team1_trap = teamPos.loc[schedule["team1"].iloc[i]]["Climb and Trap%"]

        team2_auto = teamPos.loc[schedule["team2"].iloc[i]]["Avg Auto Pieces"]
        team2_tele_speaker = teamPos.loc[schedule["team2"].iloc[i]]["Avg Teleop Speaker"]
        team2_tele_amp = teamPos.loc[schedule["team2"].iloc[i]]["Avg Teleop Amp"]
        team2_shuttle = teamPos.loc[schedule["team2"].iloc[i]]["Avg Shuttle"]
        team2_climb = teamPos.loc[schedule["team2"].iloc[i]]["Climb OnStage %"]
        team2_trap = teamPos.loc[schedule["team2"].iloc[i]]["Climb and Trap%"]

        team3_auto = teamPos.loc[schedule["team3"].iloc[i]]["Avg Auto Pieces"]
        team3_tele_speaker = teamPos.loc[schedule["team3"].iloc[i]]["Avg Teleop Speaker"]
        team3_tele_amp = teamPos.loc[schedule["team3"].iloc[i]]["Avg Teleop Amp"]
        team3_shuttle = teamPos.loc[schedule["team3"].iloc[i]]["Avg Shuttle"]
        team3_climb = teamPos.loc[schedule["team3"].iloc[i]]["Climb OnStage %"]
        team3_trap = teamPos.loc[schedule["team3"].iloc[i]]["Climb and Trap%"]

        blueNotes = team1_auto + team1_tele_speaker + team1_tele_amp + team2_auto + team2_tele_speaker + team2_tele_amp + team3_auto + team3_tele_speaker + team3_tele_amp
        blueClimb = 3*team1_climb + 5*team1_trap + 3*team2_climb + 5*team2_trap + 3*team3_climb + 5*team3_trap

        bluePoints = 0
        if (team1_auto > 1):
            bluePoints +=2
        if (team2_auto > 1):
            bluePoints +=2
        if (team3_auto > 1):
            bluePoints +=2
        
        if (team1_tele_amp + team2_tele_amp + team3_tele_amp) < 1:
            totalAmpedSpeakerPossibleBlue = 0
        else:
            totalAmpedSpeakerPossibleBlue = math.floor((team1_tele_amp + team2_tele_amp + team3_tele_amp - 1)/2)*4
        totalSpeakerNotesBlue = team1_tele_speaker + team2_tele_speaker + team3_tele_speaker
        if totalSpeakerNotesBlue >= totalAmpedSpeakerPossibleBlue:
            bluePoints += 5*totalAmpedSpeakerPossibleBlue + 2*(totalSpeakerNotesBlue - totalAmpedSpeakerPossibleBlue)
        else:
            bluePoints += 5*totalSpeakerNotesBlue

        bluePoints += 5*team1_auto + team1_tele_amp + 5*team2_auto + team2_tele_amp + 5*team3_auto + team3_tele_amp + blueClimb



        team4_auto = teamPos.loc[schedule["team4"].iloc[i]]["Avg Auto Pieces"]
        team4_tele_speaker = teamPos.loc[schedule["team4"].iloc[i]]["Avg Teleop Speaker"]
        team4_tele_amp = teamPos.loc[schedule["team4"].iloc[i]]["Avg Teleop Amp"]
        team4_shuttle = teamPos.loc[schedule["team4"].iloc[i]]["Avg Shuttle"]
        team4_climb = teamPos.loc[schedule["team4"].iloc[i]]["Climb OnStage %"]
        team4_trap = teamPos.loc[schedule["team4"].iloc[i]]["Climb and Trap%"]

        team5_auto = teamPos.loc[schedule["team5"].iloc[i]]["Avg Auto Pieces"]
        team5_tele_speaker = teamPos.loc[schedule["team5"].iloc[i]]["Avg Teleop Speaker"]
        team5_tele_amp = teamPos.loc[schedule["team5"].iloc[i]]["Avg Teleop Amp"]
        team5_shuttle = teamPos.loc[schedule["team5"].iloc[i]]["Avg Shuttle"]
        team5_climb = teamPos.loc[schedule["team5"].iloc[i]]["Climb OnStage %"]
        team5_trap = teamPos.loc[schedule["team5"].iloc[i]]["Climb and Trap%"]

        team6_auto = teamPos.loc[schedule["team6"].iloc[i]]["Avg Auto Pieces"]
        team6_tele_speaker = teamPos.loc[schedule["team6"].iloc[i]]["Avg Teleop Speaker"]
        team6_tele_amp = teamPos.loc[schedule["team6"].iloc[i]]["Avg Teleop Amp"]
        team6_shuttle = teamPos.loc[schedule["team6"].iloc[i]]["Avg Shuttle"]
        team6_climb = teamPos.loc[schedule["team6"].iloc[i]]["Climb OnStage %"]
        team6_trap = teamPos.loc[schedule["team6"].iloc[i]]["Climb and Trap%"]

        redNotes = team4_auto + team4_tele_speaker + team4_tele_amp + team5_auto + team5_tele_speaker + team5_tele_amp + team6_auto + team6_tele_speaker + team6_tele_amp
        redClimb = 3*team4_climb + 5*team4_trap + 3*team5_climb + 5*team5_trap + 3*team6_climb + 5*team6_trap

        redPoints = 0
        if (team4_auto > 1):
            redPoints +=2
        if (team5_auto > 1):
            redPoints +=2
        if (team6_auto > 1):
            redPoints +=2

        if (team4_tele_amp + team5_tele_amp + team6_tele_amp) < 1:
            totalAmpedSpeakerPossibleRed = 0
        else:
            totalAmpedSpeakerPossibleRed = math.floor((team4_tele_amp + team5_tele_amp + team6_tele_amp - 1)/2)*4
        totalSpeakerNotesRed = team4_tele_speaker + team5_tele_speaker + team6_tele_speaker
        if totalSpeakerNotesRed >= totalAmpedSpeakerPossibleRed:
            redPoints += 5*totalAmpedSpeakerPossibleRed + 2*(totalSpeakerNotesRed-totalAmpedSpeakerPossibleRed)
        else:
            redPoints += 5*totalSpeakerNotesRed

        redPoints += 5*team4_auto + team4_tele_amp + 5*team5_auto + team5_tele_amp + 5*team6_auto + team6_tele_amp + redClimb



        if (team4_tele_amp + team5_tele_amp + team6_tele_amp != 0) & (team1_tele_amp + team2_tele_amp + team3_tele_amp != 0):
            blueRP = int(blueClimb >= 9) + int(blueNotes >= 21)
            redRP = int(redClimb >= 9) + int(redNotes >= 21)
        else:
            blueRP = int(blueClimb >= 9) + int(blueNotes >= 25)
            redRP = int(redClimb >= 9) + int(redNotes >= 25)

        if(bluePoints > redPoints):
            blueRP += 2
        elif(redPoints > bluePoints):
            redRP += 2
        else:
            blueRP += 1
            redRP += 1

        rp[schedule["team1"].iloc[i]] += blueRP
        rp[schedule["team2"].iloc[i]] += blueRP
        rp[schedule["team3"].iloc[i]] += blueRP
        
        rp[schedule["team4"].iloc[i]] += redRP
        rp[schedule["team5"].iloc[i]] += redRP
        rp[schedule["team6"].iloc[i]] += redRP

        team1_arr[i] = schedule["team1"].iloc[i]
        team2_arr[i] = schedule["team2"].iloc[i]
        team3_arr[i] = schedule["team3"].iloc[i]
        team4_arr[i] = schedule["team4"].iloc[i]
        team5_arr[i] = schedule["team5"].iloc[i]
        team6_arr[i] = schedule["team6"].iloc[i]
        blueScoreArr[i] = bluePoints
        redScoreArr[i] = redPoints
        blueRPArr[i] = blueRP
        redRPArr[i] = redRP
```

```python
sched = pd.DataFrame({
    'Team1': team1_arr,
    'Team2': team2_arr,
    'Team3': team3_arr,
    'Team4': team4_arr,
    'Team5': team5_arr,
    'Team6': team6_arr,
    'BlueScore': blueScoreArr,
    'RedScore': redScoreArr,
    'BlueRP': blueRPArr,
    'RedRP': redRPArr
})

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
sched
```
